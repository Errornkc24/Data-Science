```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib .pyplot as plt
```

```python
df=pd.read_csv("Datasets-main/titanic-training-data.csv")
df.head()
```

```python
df=df.drop(["PassengerId","Name","Ticket","Cabin"],axis=1)
df.head()
```

```python
df.tail()
```

## Univeriate

```python
sns.countplot(x='Embarked',data=df)
plt.show()
```

## Bivariate

```python
sns.countplot(x='Survived', hue='Sex',data=df)
plt.show()
```

```python
sns.countplot(x='Survived', hue='Pclass',data=df)
plt.show()
```

```python
sns.countplot(x='Survived', hue='Embarked',data=df)
plt.show()
```

```python
sns.relplot(x='Age',y='Fare',data=df)
```

```python
sns.catplot(x='Survived',y='Age',data=df)
```

```python
sns.catplot(x='Pclass',y='Age',data=df, kind='box')
```

```python
sns.catplot(x='Pclass',y='Age',data=df, kind='swarm')
```

```python
sns.catplot(x='Pclass',y='Age',data=df, kind='violin')
```

```python
sns.pairplot(df)
```
